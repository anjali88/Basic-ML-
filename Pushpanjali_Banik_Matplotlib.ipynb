```
#Excerise 2.1
#Import statements
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

#hide all warnings
import warnings
warnings.filterwarnings('ignore')

#Excercise 1.1
#using the plotting function to create simple charts

x = np.arange(5) #assuming 5 students
y = (20,35,30,35,24) # student test cores
#simple bar plot
plt.bar(x,y) #Bar plot

```

```
#scatter plot
plt.scatter(x,y)
```

```
#plotting with pandas
#create histogram. line graph and the boxplot directly on dataframe
#df = pd.read_csv('Chapter_2_Code/Data/iris.csv')
from sklearn.datasets import load_wine
data = load_wine()
df = pd.DataFrame(data.data, columns = data.feature_names)
df.columns
```

```
#plot two of the column as histogrms
df[['alcohol','color_intensity']].hist()
```

```
#plot a line graph- note the key is displayed by default
df[['alcohol','color_intensity', 'total_phenols']].plot()
```

```
#plottin a box pattern
df[['alcohol','color_intensity', 'total_phenols']].boxplot()
```

```
#to plot a scatter graph, take extra step
X = df['alcohol']
y = df['color_intensity']
plt.scatter(X,y)
```
