```
#Example 1
#Generate Sample data
x = np.linspace(0, 1500, 1500)
y = np.sin(x)

# get an empty figure and add an Axes
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(1,1,1)  #row-col-num
ax.plot(x, y, 'b-', linewidth=2, label='Sample label') # line plot data on the Axes

#add title, labels and legend, etc
ax.set_ylabel('y axis label', fontsize=10)              # y label
ax.set_xlabel('x axis label', fontsize=10)              # x label
ax.legend(loc='best')                                   # legend
ax.grid(True)                                          #show grid
fig.suptitle('Sample Plot Tittle', y=1.02, fontsize=12)
fig.tight_layout(pad=1)
```

```
#Excercise 2.4 - EXAMPLES OF CUSTOMIZING PLOTS
#Example 3

#get the figure and axes all at once
fig, ax = plt.subplots(figsize=(8,4)) 
\
# plot lines 
N = 3 # the number of lines we will plot 
styles =  ['-', '--', '-.', ':'] 
markers = list('+ox') 
x = np.linspace(0, 100, 20) 
for i in range(N): # add line-by-line     
    y = x + x/5*i + i     
    s = styles[i % len(styles)]     
    m = markers[i % len(markers)]     
    ax.plot(x, y, alpha = 1, label='Line '+str(i+1)+' '+s+m,       
                  marker=m, linewidth=2, linestyle=s) 

    # add grid, legend, title and save 
ax.grid(True) 
ax.legend(loc='best', prop={'size':'large'}) 
fig.suptitle('A Simple Line Plot')
```

```
#Example 4
#get the data
N = 6
labels = list('ABCDEF')
data = np.array(list(range(N))) + np.random.rand(N)

#plot the data
fig, ax = plt.subplots(figsize=(8, 4.0))
width = 1.0
tickLocations = np.arange(N)
rectLocations = tickLocations-(width/2.0)

#for color either HEX value of the name of the color can be used
ax.bar(rectLocations, data, width, color='orange', edgecolor='#1f10ed', linewidth=4.0)

#tidy the plot
ax.set_xticks(ticks= tickLocations)
ax.set_xticklabels(labels)
ax.set_xlim(min(tickLocations)-0.5, max(tickLocations)+0.6)
ax.set_ylim(0,N)
ax.set_ylabel('y axis label', fontsize=8) #y label
ax.set_xlabel('x axis label', fontsize=8) #x label
ax.yaxis.grid(True)

fig.suptitle("Bar Plot") # add title
fig.tight_layout(pad=2)
```

```
#Example 5
# generate data
N = 5
labels = list('UVWXXY')
data = np.array(list(range(N))) + np.random.rand(N)

#plot data
fig, ax = plt.subplots(figsize=(4, 4.5))
width = 0.2
tickLocations = np.arange(N)
rectLocations = tickLocations-(width/1.5)

#for colour either HEX value value of colour can be used
ax.barh(rectLocations, data, width, color='red')

#tidy the plot
ax.set_xticks(ticks= tickLocations)
ax.set_xticklabels(labels)
ax.set_xlim(min(tickLocations)-0.5, max(tickLocations)+0.6)
ax.xaxis.grid(True)
ax.set_ylabel('y axis label', fontsize=8) #y label
ax.set_xlabel('x axis label', fontsize=8) #x label

fig.suptitle("Bar Plot1") # add title
fig.tight_layout(pad=3)
```

```
#Example 6
#generate sample data
pre = np.array([19, 6, 11, 9,13])
post = np.array([12, 11, 10, 6, 1])
labels=['Survey'+x for x in list('ABCDE')]

#the plot -left to right
fig, ax = plt.subplots(figsize=(8,4))
width = 0.3 #bar width
xlocs =np.arange(len(pre))
ax.bar(xlocs-width, pre, width, color='yellow', label='True')
ax.bar(xlocs, post, width, color='#1f10ed', label='False')

# add labels, grid and title
ax.set_xticks(ticks=list(range(len(pre))))
ax.set_xticklabels(labels)
ax.yaxis.grid(True)
ax.legend(loc='best')
ax.set_ylabel('Count')
fig.suptitle('SampleChart', y=1)
fig.tight_layout(pad=2)
```
